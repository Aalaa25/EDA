```python
def convert_to_numeric(value):
    if isinstance(value, str):
        value = value.strip().lower() 
        if 'm' in value:
            return float(value.replace('m', '').replace(',', '')) * 1_000_000
        elif 'k' in value:
            return float(value.replace('k', '').replace(',', '')) * 1_000
        elif 'b' in value:
            return float(value.replace('b', '')) * 1_000_000_000
    return value
df = df.applymap(convert_to_numeric)
```

pie chart for the largest 10 countries up to population

```python
dfpastReset = dfpast.reset_index()
dfpastMelted = dfpastReset.melt(id_vars=['country'], var_name='Year', value_name='Population')
countries = dfpastMelted.groupby('country')['Population'].sum().nlargest(10).index
dfpast10 = dfpastMelted[dfpastMelted['country'].isin(countries)]
countryPopulation = dfpast10.groupby('country')['Population'].sum()
explode = [0.1 if country == countryPopulation.idxmax() else 0 for country in countryPopulation.index]
plt.figure(figsize=(9, 9))
countryPopulation.plot(kind='pie', autopct='%1.1f%%', startangle=90, explode=explode)
plt.title("Population Distribution for Top 10 Countries")
plt.ylabel('')
plt.show()
```
